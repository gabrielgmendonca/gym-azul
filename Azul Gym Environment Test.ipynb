```python
steps = 100
total_reward = 0
with AzulEnv() as env:
    print(env.observation_space)
    for i in range(steps):
        env.render()
        obs, reward, done, info = env.step(env.action_space.sample())
        total_reward += reward
        if done:
            break
total_reward
```
